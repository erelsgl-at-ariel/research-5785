## Applying a function to all rows

```python
def to_lower_case(row):
    return row['Country or region'].lower()  # return only the country, and in lower-case

df_lower = df.apply(to_lower_case, axis=1)  # Apply the function row-by-row
df_lower
```

```python
def duplicate_country(col):
    return col[1]+col[1]  # return only the country, and duplicate it

df_test = df.apply(duplicate_country, axis=0)  
df_test
```

```python
df['Country or region'] = df_lower   # update the original DataFrame
df.head(10)
```

```python
# Adding a new column
df['Country or region Uppercase'] = df.apply(lambda row: row['Country or region'].upper(), axis=1)
df
```

```python
# Adding a new column
df['Support-by-generosity'] = df.apply(lambda row: row['Social support']/(row['Generosity']+0.001), axis=1)
df
```

```python
# The following code swaps the columns' contents, not their position:
temp = df['Support-by-generosity']
df['Support-by-generosity'] = df['Score']
df['Score'] = temp

# The following code also swaps their position:
df = df.rename(columns={"Support-by-generosity": "Score", "Score": "Support-by-generosity"})
df
```

## Grouping

```python
import numpy as np
df["RoundedScore"] = np.round(df["Score"])
df.head(10)
```

```python
df.groupby(["RoundedScore"]).count()
```

```python
df.groupby(["RoundedScore"])[["Score", "Social support", "Generosity"]].mean()
```

```python
df.groupby(["RoundedScore"])[["Score", "Social support", "Generosity"]].max()
```
